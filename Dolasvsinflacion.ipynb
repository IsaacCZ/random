```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
%matplotlib inline
```

```python
df = pd.read_excel("info.xlsx")
df.set_index("fecha", inplace = True)
dolar = pd.read_excel("dolar.xlsx")
dolar.set_index("fecha", inplace = True)
```

```python
df.plot(figsize=(25,6))
```

```python
seasonal_decompose(df["TC"]).plot()
```

```python
seasonal_decompose(df["Inflacion"]).plot()
```

```python
dolar.plot()
```

```python
seasonal_decompose(dolar["infla"]).plot()
```

```python
seasonal_decompose(dolar["tc"]).plot()
```

```python
#Bollinger Band
dolar['Close: 30 Day Mean'] = dolar['tc'].rolling(window=30).mean()
dolar['Upper'] = dolar['Close: 30 Day Mean'] + 2*dolar['tc'].rolling(window=20).std()
dolar['Lower'] = dolar['Close: 30 Day Mean'] - 2*dolar['tc'].rolling(window=20).std()
dolar[['tc','Close: 30 Day Mean','Upper','Lower']].plot(figsize=(16,6))
```

```python
#Bollinger Band
dolar['Close: 30 Day Mean'] = dolar['infla'].rolling(window=30).mean()
dolar['Upper'] = dolar['Close: 30 Day Mean'] + 2*dolar['infla'].rolling(window=20).std()
dolar['Lower'] = dolar['Close: 30 Day Mean'] - 2*dolar['infla'].rolling(window=20).std()
dolar[['infla','Close: 30 Day Mean','Upper','Lower']].plot(figsize=(16,6))
```
